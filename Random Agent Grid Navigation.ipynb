```python
import random

def run_simulation(grid_size, max_steps):
    x, y = 0, 0
    goal = (grid_size - 1, grid_size - 1)
    
    for step in range(1, max_steps + 1):
        move = random.choice(['up', 'down', 'left', 'right'])
        
        # Movement logic with boundary checks
        if move == 'up' and y < grid_size - 1:
            y += 1
        elif move == 'down' and y > 0:
            y -= 1
        elif move == 'right' and x < grid_size - 1:
            x += 1
        elif move == 'left' and x > 0:
            x -= 1
            
        if (x, y) == goal:
            return f"SUCCESS: Goal reached in {step} steps."
            
    return f"FAILURE: Goal not reached within {max_steps} steps."

# --- Experiment 1: 10x10 Grid ---
print("--- Grid Size 10 (Goal: 9,9) ---")
for s in [20, 30, 40, 50, 100]:
    print(f"Limit {s}: {run_simulation(10, s)}")

# --- Experiment 2: 21x21 Grid ---
print("\n--- Grid Size 21 (Goal: 20,20) ---")
for s in [100, 200, 500, 1000, 2000]:
    print(f"Limit {s}: {run_simulation(21, s)}")
```
